```python
import cv2
import numpy as np

cap= cv2.VideoCapture('family.mp4')
ret,frame1=cap.read()
ret,frame2=cap.read()

while cap.isOpened():
    diff= cv2.absdiff(frame1,frame2)
    gray = cv2.cvtColor(diff,cv2.COLOR_BGR2GRAY) # gray
    # it's easier to find contours in grayscale than color mode
    blur = cv2.GaussianBlur(gray,(5,5),0)
    _,thresh = cv2.threshold(blur,20,255,cv2.THRESH_BINARY)
#     dilated = cv2.dilation(thresh,None,iterations=3)
    dilated = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,(5,5),iterations=8)
    _,contours,_ = cv2.findContours(dilated,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    # remove noise
    
    
    
#     cv2.drawContours(frame1,contours,-1,(0,255,0),2) # draw contour on orignal image
    # draw rectangle instead of drawing contour 
    
    
    for contour in contours:
        (x,y,w,h) = cv2.boundingRect(contour)
        # if less than xx values, drop it(xx is defined according to the objects)
        if cv2.contourArea(contour) < 8000:
            continue   # not draw rectangle
        cv2.rectangle(frame1,(x,y),(x+w,y+h),(0,255,0),2)
        cv2.putText(frame1,"status:{}".format('Movement'),(10,20),cv2.FONT_HERSHEY_SIMPLEX,
                   1,(0,0,255),3)
        print()
    cv2.imshow('feed',frame1)
    frame1=frame2
    ret,frame2=cap.read()
    
#     cv2.imshow('dilated',dilated)
#     cv2.imshow('thresh',thresh)
    
    if cv2.waitKey(40) == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
